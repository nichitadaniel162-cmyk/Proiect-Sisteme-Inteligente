```python
# Importam bibliotecile necesare
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_openml

# 1. Incarcarea datelor
print("Incarcam setul de date MNIST. Te rog ai rabdare (poate dura 1-2 minute)...")
mnist = fetch_openml('mnist_784', version=1, as_frame=True, parser='auto')

X = mnist.data
y = mnist.target.astype(int)
print("Datele au fost incarcate cu succes!\n")

# 2. Analiza Exploratorie (EDA) si Curatare
print("--- Analiza Exploratorie ---")
print(f"Dimensiunea setului de caracteristici (X): {X.shape}")
print(f"Dimensiunea setului de etichete (y): {y.shape}")
print("Avem 70.000 de imagini, fiecare imagine avand 784 de pixeli.\n")

valori_lipsa = X.isnull().sum().sum()
print(f"Numarul total de valori lipsa in baza de date: {valori_lipsa}")
if valori_lipsa == 0:
    print("Baza de date este perfect curata. Nu este nevoie de imputare de date lipsa.\n")

# 3. Vizualizarea datelor
print("Generam graficele...")

# Distributia claselor (cate imagini avem pentru fiecare cifra)
plt.figure(figsize=(8, 4))
sns.countplot(x=y, palette='viridis')
plt.title('Distributia cifrelor in setul de date MNIST')
plt.xlabel('Cifra')
plt.ylabel('Numar de aparitii')
plt.show()

# Afisarea catorva imagini reale pentru a vedea cum arata datele
print("\nVizualizam cateva exemple de imagini reale din setul de date:")
plt.figure(figsize=(10, 4))
for i in range(10):
    plt.subplot(2, 5, i + 1)

    # Transformam randul de 784 pixeli inapoi in format de imagine 28x28
    imagine_aplatizata = X.iloc[i].values
    imagine_2d = imagine_aplatizata.reshape(28, 28)

    plt.imshow(imagine_2d, cmap='gray')
    plt.title(f"Eticheta: {y.iloc[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()

print("Analiza pentru saptamana 4 a fost finalizata!")
# 4. Grafice suplimentare: Imaginile medii
print("\nGeneram imaginile medii pentru fiecare cifra (cum arata 'media' tuturor imaginilor):")
plt.figure(figsize=(12, 5))
for cifra in range(10):
    # Selectam doar randurile corespunzatoare cifrei curente
    indici = (y == cifra)
    imagini_cifra = X[indici]

    # Calculam media pentru fiecare pixel
    imagine_medie = imagini_cifra.mean(axis=0).values.reshape(28, 28)

    plt.subplot(2, 5, cifra + 1)
    # Folosim o paleta de culori 'magma' ca sa arate spectaculos
    plt.imshow(imagine_medie, cmap='magma')
    plt.title(f"Media cifrei: {cifra}")
    plt.axis('off')

plt.tight_layout()
plt.show()


# 5. Grafice suplimentare: Harta variantei (Ce pixeli contin cea mai multa informatie?)
print("\nGeneram Harta Variantei (Heatmap) pentru tot setul de date:")
# Calculam varianta fiecarui pixel in parte pe toata baza de date
varianta_pixeli = X.var(axis=0).values.reshape(28, 28)

plt.figure(figsize=(6, 5))
sns.heatmap(varianta_pixeli, cmap='viridis', cbar=True, xticklabels=False, yticklabels=False)
plt.title("Harta Variantei Pixelilor")
plt.show()

print("\nConcluzie vizuala importanta pentru proiect:")
print("Zonele inchise la culoare (marginile) au varianta zero (sunt mereu negre).")
print("Acest lucru justifica perfect necesitatea utilizarii PCA in etapa urmatoare,")
print("pentru a elimina aceste dimensiuni redundante si a pastra doar centrul imaginii!")
```

```python
# importuri pt saptamana 6
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.ensemble import RandomForestClassifier

print("--- Rulare saptamana 6: Feature engineering ---")

# 1. split si esantionare (iau doar un subset sa nu crape colab-ul la t-SNE)
# din totalul de 70k de imagini opresc 10k pentru antrenare ca sa mearga mai repede
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=10000, test_size=2000, random_state=42, stratify=y)

# standardizare (necesar inainte sa aplic pca)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
print("1. split si scalare finalizate.")

# 2. feature importance
print("2. calculez importanta pixelilor cu un random forest...")
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X_train_scaled, y_train)

# refac scorurile sub forma de imagine 28x28 ca sa aiba sens vizual
importanta_pixeli = rf.feature_importances_.reshape(28, 28)

plt.figure(figsize=(6, 5))
sns.heatmap(importanta_pixeli, cmap='hot', cbar=True, xticklabels=False, yticklabels=False)
plt.title("Feature Importance\n(zonele deschise arata pixelii care conteaza cel mai mult pt clasificare)")
plt.show()

# 3. corelatii pe pca
print("3. fac matricea de corelatie pt primele 10 componente...")
pca_viz = PCA(n_components=10, random_state=42)
X_pca_10 = pca_viz.fit_transform(X_train_scaled)

# bag datele intr-un dataframe sa pot apela direct .corr()
df_pca = pd.DataFrame(X_pca_10, columns=[f"PC{i+1}" for i in range(10)])
corelatii_pca = df_pca.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corelatii_pca, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title("Corelatii primele 10 componente PCA\n(0 peste tot = variabilele sunt complet independente acum)")
plt.show()

# 4. comparatie PCA vs t-SNE in 2D
print("4. rulez pca si t-sne pe 2 dimensiuni (dureaza putin t-sne-ul)...")
# aplicare t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne_2d = tsne.fit_transform(X_train_scaled)

# aplicare PCA
pca_2d = PCA(n_components=2, random_state=42)
X_pca_2d = pca_2d.fit_transform(X_train_scaled)

# plotare in paralel
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# scatter pt PCA
scatter1 = ax1.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=y_train, cmap='tab10', alpha=0.7, s=10)
ax1.set_title("Rezultat PCA (metoda liniara)")
fig.colorbar(scatter1, ax=ax1)

# scatter pt t-SNE
scatter2 = ax2.scatter(X_tsne_2d[:, 0], X_tsne_2d[:, 1], c=y_train, cmap='tab10', alpha=0.7, s=10)
ax2.set_title("Rezultat t-SNE (metoda neliniara)\n(se vede clar cum a separat cifrele pe grupuri)")
fig.colorbar(scatter2, ax=ax2)

plt.show()
```
