```python
# Importam bibliotecile necesare
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_openml

# 1. Incarcarea datelor
print("Incarcam setul de date MNIST. Te rog ai rabdare (poate dura 1-2 minute)...")
mnist = fetch_openml('mnist_784', version=1, as_frame=True, parser='auto')

X = mnist.data
y = mnist.target.astype(int)
print("Datele au fost incarcate cu succes!\n")

# 2. Analiza Exploratorie (EDA) si Curatare
print("--- Analiza Exploratorie ---")
print(f"Dimensiunea setului de caracteristici (X): {X.shape}")
print(f"Dimensiunea setului de etichete (y): {y.shape}")
print("Avem 70.000 de imagini, fiecare imagine avand 784 de pixeli.\n")

valori_lipsa = X.isnull().sum().sum()
print(f"Numarul total de valori lipsa in baza de date: {valori_lipsa}")
if valori_lipsa == 0:
    print("Baza de date este perfect curata. Nu este nevoie de imputare de date lipsa.\n")

# 3. Vizualizarea datelor
print("Generam graficele...")

# Distributia claselor (cate imagini avem pentru fiecare cifra)
plt.figure(figsize=(8, 4))
sns.countplot(x=y, hue=y, palette='viridis', legend=False)
plt.title('Distributia cifrelor in setul de date MNIST')
plt.xlabel('Cifra')
plt.ylabel('Numar de aparitii')
plt.show()

# Afisarea catorva imagini reale pentru a vedea cum arata datele
print("\nVizualizam cateva exemple de imagini reale din setul de date:")
plt.figure(figsize=(10, 4))
for i in range(10):
    plt.subplot(2, 5, i + 1)

    # Transformam randul de 784 pixeli inapoi in format de imagine 28x28
    imagine_aplatizata = X.iloc[i].values
    imagine_2d = imagine_aplatizata.reshape(28, 28)

    plt.imshow(imagine_2d, cmap='gray')
    plt.title(f"Eticheta: {y.iloc[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()

print("Analiza pentru saptamana 4 a fost finalizata!")
```
